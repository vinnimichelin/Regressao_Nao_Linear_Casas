```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
house = pd.read_excel('house.xlsx')

house.head()
```

```python
house.info()
```

```python
#Transformando em dummies as colunas de texto.

house = pd.get_dummies(house)

house.head()
```

```python
#Dropando as linhas vazias

house = house.dropna()
```

```python
#Fazendo as variaveis X e Y

labels = np.array(house['SalePrice'])

features = house.drop('SalePrice', axis=1)

features_list = list(house.columns)

features = np.array(features)
```

```python
#Dividindo a base em treino e teste

train_features, test_features, train_labels,test_labels = train_test_split(features,labels, test_size=0.25, random_state=42)
```

```python
#Treinando o RandomForestRegressor

rf = RandomForestRegressor(n_estimators=1000, random_state=42)
rf.fit(train_features, train_labels)
```

```python
#Executando o RandomForestRegressor

prediction_rf = rf.predict(test_features)

error_rf = abs(prediction_rf - test_labels)

r_sq = rf.score(features, labels)

print('R^2:', r_sq)
print('MAE:', metrics.mean_absolute_error(test_labels, prediction_rf))
print('MSE:', metrics.mean_squared_error(test_labels, prediction_rf))
```

```python
#Treinando o AdaBoostRegressor

ada = AdaBoostRegressor(n_estimators=100)

ada.fit(train_features, train_labels)

ada_pred = ada.predict(test_features)
```

```python
#Executando o AdaBoostRegressor

error_ada = abs(ada_pred - test_labels)

r_sq = ada.score(features, labels)

print('R^2:', r_sq)
print('MAE:', metrics.mean_absolute_error(test_labels, ada_pred))
print('MSE:', metrics.mean_squared_error(test_labels, ada_pred))
```

```python
#Treinando o GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=100)

gbr.fit(train_features, train_labels)

gbr_pred = gbr.predict(test_features)
```

```python
#Executando o GradientBoostingRegressor

error_gbr = abs(gbr_pred - test_labels)

r_sq = gbr.score(features, labels)

print('R^2:', r_sq)
print('MAE:', metrics.mean_absolute_error(test_labels, gbr_pred))
print('MSE:', metrics.mean_squared_error(test_labels, gbr_pred))
```

# **Após os testes, podemos ver que o GradientBoostingRegressor é o método que obteve o melhor desempenho.**
